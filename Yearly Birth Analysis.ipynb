# ANALYSIS OF YEARLY BIRTH RECORDS BY COUNTRIES

### The project is aimed at determing the yearly births of countries as presented in a dataset, also visualising the birth records with respect to year taken.
### 3 countries were selected for analysis purpose, United States of America, Ghana and Netherlands. Same techniques applied for these 3 countries can be applied in other countries to get insights in their records.
### Dataset was sourced from Kaggle.com

```python
# Importing libraries

import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# loading data

data = pd.read_csv("C:/Users/USER/Desktop/screened data/BirthMonthsData.csv")
```

### Analysis of records for the United States of America

```python
USA = data[data["Country or Area"] == "United States of America"]
```

```python
USA
```

```python
USA.loc[(USA.Month == "January") & (USA.Year == 2019)]
```

```python
USA_TOTAL  = USA.loc[USA['Month'] == 'Total']
USA_TOTAL
```

```python
USA_TOTAL.describe()
```

#### Maximum birth in USA from records is over 4.26mil
#### Minimum birth in USA From records is around 3.14mil
#### Mean birth is 3.83mil
#### Third quartile of the births is 4.02mil 

```python
# Total births in United States of America

USA_TOTAL['Number_of_Births'].sum()
```

```python
plt.figure(figsize= (12,6))
plt.plot(USA_TOTAL.Year, USA_TOTAL["Number_of_Births"])
plt.xlabel = "Years"
plt.ylabel = "Birth in millions"
plt.title = "Total births by Year for USA"
plt.show()
```

#### From the plot above, the common deduction is that maximum birth was recorded in 2008 while the least was recorded in 1973.
#### High birth rate was recored in late 1980s which peaked in 1990.
#### There was a surge in birth rate from early 2000s which peaked in 2008 and has been on a continuous reduction process to the last record

### Analysis of records for Ghana

```python
GH_TOTAL = GH.loc[GH["Month"] != "Total"]
GH_TOTAL
```
